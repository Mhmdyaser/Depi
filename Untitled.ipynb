```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.feature_selection import SelectPercentile, f_classif
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, classification_report, confusion_matrix,
                             roc_curve, auc, roc_auc_score)
import warnings
warnings.filterwarnings('ignore')
```

```python
# 1. Load dataset
df = pd.read_csv('CardionDataset.csv')
print("Original shape:", df.shape)
```

```python
df.head()
```

```python
# Drop 'id' column if present
if 'id' in df.columns:
    df.drop('id', axis=1, inplace=True)
```

```python
# Convert age from days to years
df['age_years'] = df['age'] / 365.25
df.drop('age', axis=1, inplace=True)
```

```python
df.head()
```

```python
# Outlier removal (optional but recommended)

def remove_outliers(df):
    # Blood pressure
    df = df[(df['ap_hi'] > 50) & (df['ap_hi'] < 250)]
    df = df[(df['ap_lo'] > 30) & (df['ap_lo'] < 200)]
    df = df[df['ap_hi'] >= df['ap_lo']]
    # Height and weight
    df = df[(df['height'] >= 100) & (df['height'] <= 220)]
    df = df[(df['weight'] >= 30) & (df['weight'] <= 200)]
    return df

df = remove_outliers(df)
print("After outlier removal:", df.shape)
```

```python
df.head()
```

```python
# Feature engineering: BMI
df['bmi'] = df['weight'] / ((df['height']/100) ** 2)
df = df[(df['bmi'] > 10) & (df['bmi'] < 50)]
```

```python

# Numerical continuous features
numerical_cols = ['age_years', 'height', 'weight', 'ap_hi', 'ap_lo', 'bmi']
# Categorical features 
categorical_cols = ['gender', 'cholesterol', 'gluc', 'smoke', 'alco', 'active']

X = df[categorical_cols + numerical_cols]
y = df['cardio']
```

```python
#  Preprocessing pipeline
# impute with median, then scale
numerical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

#  impute with most frequent, then one-hot encode
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(drop='first', sparse_output=False))  # avoid dummy trap
])

# Combine both pipelines
preprocessor = ColumnTransformer(transformers=[
    ('num', numerical_transformer, numerical_cols),
    ('cat', categorical_transformer, categorical_cols)
])
```

```python
# 5. Feature selection (SelectPercentile)
# Transform the data
X_processed = preprocessor.fit_transform(X)
print("Shape after preprocessing (including one-hot encoding):", X_processed.shape)

# Apply SelectPercentile (keep top 50% features by default)
selector = SelectPercentile(score_func=f_classif, percentile=50)
X_selected = selector.fit_transform(X_processed, y)
print("Shape after feature selection:", X_selected.shape)

# Get selected feature indices (optional)
selected_indices = selector.get_support(indices=True)
```

```python
# 6. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42, stratify=y)
```

```python
# 7. Models to evaluate
models = {
    'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000),
    'K-Nearest Neighbors': KNeighborsClassifier(),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100),
    #'Support Vector Machine': SVC(probability=True, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(random_state=42),
}
```

```python
# 8. Train and evaluate

results = {}
print("\n" + "="*60)
print("MODEL EVALUATION (after preprocessing + feature selection)")
print("="*60)

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1] if hasattr(model, "predict_proba") else None
    
    acc = accuracy_score(y_test, y_pred)
    results[name] = {'accuracy': acc, 'model': model, 'y_pred': y_pred, 'y_prob': y_prob}
    
    print(f"\n{name}")
    print(f"Accuracy: {acc:.4f}")
    print(classification_report(y_test, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()
```

```python
# 9. ROC Curves
# ---------------------------
plt.figure(figsize=(10,8))
for name, res in results.items():
    if res['y_prob'] is not None:
        fpr, tpr, _ = roc_curve(y_test, res['y_prob'])
        roc_auc = auc(fpr, tpr)
        plt.plot(fpr, tpr, label=f'{name} (AUC = {roc_auc:.3f})')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves')
plt.legend(loc='lower right')
plt.show()
```

```python
# 10. Feature importance (Random Forest)

onehot_features = preprocessor.named_transformers_['cat'].named_steps['onehot'].get_feature_names_out(categorical_cols)
all_features = numerical_cols + list(onehot_features)
selected_features = [all_features[i] for i in selected_indices]

rf_model = results['Random Forest']['model']
importances = rf_model.feature_importances_
feat_imp_df = pd.DataFrame({'feature': selected_features, 'importance': importances}).sort_values('importance', ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='importance', y='feature', data=feat_imp_df.head(15))
plt.title('Top 15 Feature Importances (Random Forest)')
plt.tight_layout()
plt.show()
```

```python
# 11. Additional visualizations

# Age distribution by target
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='age_years', hue='cardio', kde=True, bins=30)
plt.title('Age Distribution by Heart Disease')
plt.show()
```

```python
# Blood pressure boxplots
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.boxplot(x='cardio', y='ap_hi', data=df)
plt.title('Systolic BP vs Target')
plt.subplot(1,2,2)
sns.boxplot(x='cardio', y='ap_lo', data=df)
plt.title('Diastolic BP vs Target')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap (original numeric features)
plt.figure(figsize=(10,8))
numeric_df = df[numerical_cols + ['cardio']]
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap (Numerical Features)')
plt.show()
```

```python
# 12. Best model summary
best_model = max(results, key=lambda x: results[x]['accuracy'])
print(f"\nBest model: {best_model} with accuracy {results[best_model]['accuracy']:.4f}")
```
